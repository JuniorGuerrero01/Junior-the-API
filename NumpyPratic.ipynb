````python
## Operaciones con vectores
## 1. Crea dos vectores de 5 elemento cada uno
## 2. Haz su suma,resta,multi elemento a elemento (Hadamard) y la multiplicacion con dot.

import numpy as np

x = np.array([9,7,5,6,3])
y = np.array([8,3,5,6,2])

resta = x-y
suma = x+y

multi_hadamard = x*y
multi_dot = np.dot(x,y)

print(f"La resta del Vector es: ", resta,f"La suma de los Vectores es: ",suma)
print(f"La multiplicacion del hadamard es: ",multi_hadamard)
print(f"Esta es la multiplicacion del dot: ",multi_dot)
````

````python
## Estadistica Basica
## 1. Crea un array con 15 numeros aleatorio del 1 al 50
## 2. Calcula el promedio, el maximo y el minimo.
import numpy as np

np.random.seed(42)
aleatorio = np.random.randint(1,51, size=15)
print(aleatorio)

promedio = aleatorio.mean()
minimo = aleatorio.min()
maximo = aleatorio.max()

print(f"El promedio es de: ",promedio)
print(f"El minimo es de: ",minimo)
print(f"El maximo es de: ",maximo)
````

````python
## Grafico simple
## 1. Genera 50 valores entre 0 y 2*np.pi con linspace
## 2. Calcula su seno
## 3. haz un grafico de la funcion seno usando matplotlib


import numpy as np
import matplotlib.pyplot as plt

Num = np.linspace(0,2*np.pi)
seno = np.sin(Num)

%matplotlib inline

plt.figure(figsize=(10,6))
plt.plot(Num,seno)
plt.title("GRAFICA DEL SENO")
plt.xlabel("x")
plt.xlabel("sin(x)")
plt.grid(True)
plt.show()







````

````python
## Suma y multi de matrices
## 1. Crea dos matrices 2x2 con valores enteros
## 2. Haz su suma
## 3. Multiplica una de ellas por un escalar ejem: 5
## 4. Multiplica las dos matrices con el operador @.


import numpy as np

x = np.array([[2,4],[6,3]])
y = np.array([[5,7],[9,1]])

print(f"{x}\n")
print(y)

suma = x+y

print(f"suma de las dos matrices: ")
suma = x+y
print(suma)

print(f"La multiplicacion es: ")
multi = x*5
print(multi)

print(f"Multiplicando las dos matrices con @")
multi_2 = x@y

print(multi_2)


````

````python
## Producto hadamard vs Multiplicacion de matrices
## 1. Crea dos matrices 3x3
## 2. Haz el producto elemento a elemento hadamard
## 3. Haz la multiplicacion de matrices con dot
## 4. Explica la diferencia del resultado

import numpy as np

x =  np.array([[2,4,6],[6,3,2], [3,5,7]])
y =  np.array([[5,7,2],[9,1,5],[4,5,5]])

print(f"{x}\n")
print(y)

multi = x*y

print(f"La multiplicacion es: ")
print(multi)

multi1 = np.dot(x,y)
print(f"La multiplicacion por hadamard: ",)
print(multi1)

##explicaci√≥n en:

# Multiplicaci√≥n Hadamard (elemento a elemento)
# Cada elemento de un array se multiplica por el elemento correspondiente del otro array.
# Resultado: un nuevo array del mismo tama√±o que los originales.
# Ejemplo conceptual: [a, b, c] * [x, y, z] => [a*x, b*y, c*z]

# Producto punto (dot)
# Multiplica los elementos correspondientes y luego suma todos esos productos.
# Resultado: un solo n√∫mero (escalar).
# Ejemplo conceptual: [a, b, c] . [x, y, z] => a*x + b*y + c*z

# Resumen:
# Hadamard -> devuelve array
# Dot -> devuelve n√∫mero
```

````
